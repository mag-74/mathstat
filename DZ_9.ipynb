Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp-признак), а за y-значения скорингового балла (то есть, ks - это целевая переменная). Произвести расчет как с использованием intercept, так и без.


```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
plt.scatter(zp, ks)
```

```python
b1 = (np.mean(ks*zp)-np.mean(zp)*np.mean(ks))/(np.mean(zp**2)-np.mean(zp)**2)
b1
```

```python
b0 = np.mean(ks) - b1 * np.mean(zp)
b0
```

```python
plt.scatter(zp, ks)
plt.plot(ks, b1*ks+b0)
```

ЗАДАЧА №2. Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).

```python
n = len(zp)
m = len(ks)
n, m
```

```python
# Зададим скорость обучения (играя этим значением, добился чего-то вразумительного при расчете, но ...)
alpha = 1e-12
B1 = 1
```

```python
# Зададим функцию для расчета среднеквадратичной ошибки:
def mse_(B1, y=ks, x=zp, n=10):
    return np.sum((B1 * x - y)**2)/n
```

```python
# Запускаем метод градиентного спуска:
for i in range (300000):
    B1 -= alpha * (2/n) * np.sum((B1 * zp - ks) * zp)
    if i % 10000 == 0:
        print('Итерация={i}, B1={B1}, mse={mse}'.format(i=i, B1=B1, mse=mse_(B1)))
```
